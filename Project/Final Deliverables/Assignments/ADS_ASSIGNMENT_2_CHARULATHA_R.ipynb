**Find outliers using statistical methods**

```python
def find_outliers_IQR(ads):
  q1=ads.quantile(0.25)
  q3=ads.quantile(0.75)
  IQR=q3-q1
  out = ads[((ads<(q1-1.5*IQR)) | (ads>(q3+1.5*IQR)))]
  return out
```

```python
out = find_outliers_IQR(ads['bathrooms'])
print('number of outliers:'+ str(len(out)))
print('max outlier value:'+ str(out.max()))
print('min outlier value:'+ str(out.min()))
out
```

**Replace outliers using imputation as if they were missing values.**

```python
def impute_outliers_IQR(ads):
   q1=ads.quantile(0.25)
   q3=ads.quantile(0.75)
   IQR=q3-q1
   upper = ads[~(ads>(q3+1.5*IQR))].max()
   lower = ads[~(ads<(q1-1.5*IQR))].min()
   ads = np.where(ads > upper,
       ads.mean(),
       np.where(
           ads < lower,
           ads.mean(),
           ads
           )
       )
   return ads
```

```python
ads['bedrooms'] = impute_outliers_IQR(ads['bedrooms'])
ads.describe()['bedrooms']
```
